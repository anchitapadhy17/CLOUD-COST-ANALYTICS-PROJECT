```python
import csv
import random
from datetime import datetime, timedelta

def generate_cloud_dataset(filename="cloud_usage_raw.csv", num_rows=2500):
    accounts = ["acct-009", "ACCT-010", " acct-011 ", "ACCT-012", "Acct-013"]
    services = ["Compute", "Storage", "Database", "Networking", "SaaS"]
    skus = {
        "Compute": ["VM-std_4", "VM-STD-4", "compute-v2", "GPU-inst"],
        "Storage": ["S3-Standard", "s3-standard", "Cold-Archive", "ebs-vol"],
        "Database": ["RDS-MultiAZ", "Aurora-DB", "AURORA-DB", "NoSQL-Tbl"],
        "Networking": ["DataTransfer", "LoadBalancer", "VPN-Gateway"],
        "SaaS": ["Email-API", "Auth-Service", "Monitoring-App"]
    }
    units = ["sec", "seconds", "min", "hrs", "GB", "units"]
    regions = ["ap-south 1", "ap-south-1", "us-east-1", "us-east 1", "eu-west-1", "global"]

    with open(filename, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["Usage_ID", "Account", "TS", "Service", "SKU", "Usage", "Unit", "Cost", "Region", "Ticket_ID"])

        for i in range(num_rows):
            usage_id = f"U{7000 + i}"
            service = random.choice(services)
            sku = random.choice(skus[service])
            
            # Introduce inconsistencies
            account = random.choice(accounts)
            region = random.choice(regions)
            unit = random.choice(units)
            
            # Timestamp Logic (occasionally create a "dirty" hour like 25:00)
            base_date = datetime(2026, 3, 10) + timedelta(hours=random.randint(0, 500))
            if random.random() < 0.05: # 5% chance of invalid time skew
                ts = base_date.strftime("%Y/%m/%d") + " 25:05"
            elif random.random() < 0.3:
                ts = base_date.strftime("%Y-%m-%d %H:%M:%S+00:00")
            else:
                ts = base_date.strftime("%Y/%m/%d %H:%M")

            # Usage and Cost logic
            usage_val = random.randint(100, 20000)
            cost_val = round(usage_val * 0.1, 2)
            
            # Formatting "dirtiness"
            usage = f'"{usage_val:,}"' if random.random() < 0.2 else str(usage_val)
            cost = f'"₹ {cost_val:,}"' if random.random() < 0.3 else str(cost_val)
            
            ticket_id = f"T-{random.randint(100, 999)}" if random.random() < 0.4 else "NULL"

            writer.writerow([usage_id, account, ts, service, sku, usage, unit, cost, region, ticket_id])
            
            # Duplicate generation (approx 10% of rows are duplicates with different casing/units)
            if random.random() < 0.1 and i < num_rows - 1:
                writer.writerow([usage_id, account.strip().upper(), ts, service, sku.upper(), 
                                 usage_val, "seconds", cost_val, region.replace(" ", "-"), ticket_id])
                i += 1 

    print(f"Successfully generated {filename} with ~2500 rows.")

generate_cloud_dataset()
```
